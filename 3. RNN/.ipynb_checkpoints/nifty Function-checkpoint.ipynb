```python
import pandas as pd
import numpy as np 

from keras.models import Sequential 
from keras.layers import LSTM, Dense, Dropout

from sklearn.preprocessing import MinMaxScaler

import matplotlib.pyplot as plt
%matplotlib inline

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
```

```python
df = pd.read_csv('nifty.csv')
cols = ['Price', 'Open', 'High', 'Low']
df[cols]= df[cols].replace(',', "", regex = True).astype(float)
df.head()

test = pd.read_csv('nifty_test.csv')
test[cols]= test[cols].replace(',', "", regex = True).astype(float)

test.head()
```

```python
train_Open = df.iloc[:, 2:3].values
train_High = df.iloc[:, 3:4].values
train_Low = df.iloc[:, 4:5].values
train_Price = df.iloc[:, 1:2].values
```

```python
test_Open = test.iloc[:, 2:3].values
test_High = test.iloc[:, 3:4].values
test_Low = test.iloc[:, 4:5].values
test_Price = test.iloc[:, 1:2].values
```

```python
def build_model(train_set, test_set):
    sc = MinMaxScaler()
    train_scale = sc.fit_transform(train_set)
    xtr = train_scale[0:train_scale.shape[0]-1]
    ytr = train_scale[1:train_scale.shape[0]]
    xtr = np.reshape(xtr, (train_scale.shape[0]-1, 1, 1))
    
    model = Sequential()
    model.add(LSTM(units=4, activation = 'sigmoid', input_shape=(None,1)))
    model.add(Dense(units=1))
    model.compile(optimizer = 'adam', loss = 'mean_squared_error')
    model.fit(xtr, ytr, epochs = 512, batch_size=64)
    test_scale = sc.transform(test_set)
    test_data = np.reshape(test_scale, (test_scale.shape[0], test_scale.shape[1], 1))
    
    predicted_stock = model.predict(test_data)
    predicted_stock = sc.inverse_transform(predicted_stock)
    
    plt.plot(test_set, color = 'red', label = 'Real Google Stock Price')
    plt.plot(predicted_stock, color = 'blue', label = 'Predicted Google Stock Price')
    plt.title('Google Stock Price Prediction')
    plt.xlabel('Time')
    plt.ylabel('Google Stock Price')
    plt.legend()
    plt.show()
    
    test_temp = pd.DataFrame(test_set)
    pred_temp = pd.DataFrame(predicted_stock, columns=['Predicted'])
    pred_total = pd.concat((test_temp, pred_temp), axis=1)
    
    print('Last Day price', test_set[-5:])
    print('Last Day Prediction', predicted_stock[-5:])
    print(pred_total)   
    
```

```python
build_model(train_High, test_High)
```
