```python
import pandas as pd
import numpy as np 

from keras.models import Sequential 
from keras.layers import LSTM, Dense, Dropout

from sklearn.preprocessing import MinMaxScaler

import matplotlib.pyplot as plt
%matplotlib inline

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
```

```python
df = pd.read_csv('nifty.csv')
df.head()
df.shape
```

```python
df.columns
```

```python
cols = ['Price', 'Open', 'High', 'Low']
```

```python
df[cols]= df[cols].replace(',', "", regex = True).astype(float)
df.head()
df.dtypes
```

```python
train_set = df.iloc[:, 2:3].values
train_set
```

```python
sc = MinMaxScaler()
train_scale = sc.fit_transform(train_set)
```

```python
train_scale.shape
```

```python
xtr = train_scale[0:train_scale.shape[0]-1]
ytr = train_scale[1:train_scale.shape[0]]

xtr = np.reshape(xtr, (train_scale.shape[0]-1, 1, 1))
```

```python
model = Sequential()
model.add(LSTM(units=4, activation = 'sigmoid', input_shape=(None,1)))
model.add(Dense(units=1))
model.compile(optimizer = 'adam', loss = 'mean_squared_error')
model.summary()
```

```python
model.fit(xtr, ytr, epochs = 500, batch_size=64)
```

```python
test = pd.read_csv('nifty_test.csv')
test.head()
```

```python
cols = ['Price', 'Open', 'High', 'Low']
```

```python
test[cols]= test[cols].replace(',', "", regex = True).astype(float)
test.head()
```

```python
test_set = test.iloc[:, 2:3].values
```

```python
test_scale = sc.transform(test_set)
```

```python
test_data = np.reshape(test_scale, (test_scale.shape[0], test_scale.shape[1], 1))
test_data.shape
```

```python
predicted_stock = model.predict(test_data)
predicted_stock = sc.inverse_transform(predicted_stock)
```

```python
plt.plot(test_set, color = 'red', label = 'Real Google Stock Price')
plt.plot(predicted_stock, color = 'blue', label = 'Predicted Google Stock Price')
plt.title('Google Stock Price Prediction')
plt.xlabel('Time')
plt.ylabel('Google Stock Price')
plt.legend()
plt.show()
```
