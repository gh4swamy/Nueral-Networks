```python
import pandas as pd, numpy as np
df = pd.read_csv('Churn_Modelling.csv')
df.head()
```

```python
df = df.iloc[:, 3:]
df.head()
```

```python
from sklearn.preprocessing import LabelBinarizer
lb = LabelBinarizer()
df['Gender'] = lb.fit_transform(df['Gender'])
df.head()
```

```python
print(df.Geography.unique())
print(df.Gender.nunique())
```

```python
# Apply Catogrocal Values tp Gorgraphy
df.Geography = pd.Categorical(df.Geography).codes
df.head()
```

```python
df.shape
```

```python
x = df.iloc[:, :10].values
y = df.iloc[:, 10].values
print(x[:5])
print(y[:5])
```

```python
from sklearn.model_selection import train_test_split
xtr, xts, ytr, yts = train_test_split(x, y, test_size=0.3, random_state = 42)
print(xtr.shape)
print(xts.shape)
print(ytr.shape)
print(yts.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
xtr = sc.fit_transform(xtr)
```

```python
xts = sc.transform(xts)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout

model = Sequential()
model.add(Dense(input_dim=10, units=6, activation = 'relu' ))
model.add(Dense(units =6, activation = 'relu'))
model.add(Dense(units=1, activation = 'sigmoid'))

model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(xtr, ytr, batch_size=200, epochs =40)
```

```python
ypr = model.predict(xts)
ypr = ypr>0.5
ypr
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(yts, ypr)
cm
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
```

```python
def build_model():
    model = Sequential()
    model.add(Dense(input_dim=10, units=6, activation = 'relu' ))
    model.add(Dense(units =6, activation = 'relu'))
    model.add(Dense(units=1, activation = 'sigmoid'))
    model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
    return model
```

```python
classifer = KerasClassifier(build_fn = build_model, batch_size=10, epochs = 10)
accuraci = cross_val_score(estimator = classifer, X= xtr, y = ytr, cv=10)
```

```python
print(accuraci.mean())
print(accuraci.std())
```

```python
from sklearn.model_selection import GridSearchCV

grid_params = {'batch_size':[10, 15, 30], 
              'epochs':[10, 20]}

classifier = KerasClassifier(build_fn = build_model)

grid_search = GridSearchCV(estimator = classifier, param_grid = grid_params, scoring = 'accuracy', cv =10)
gd_sr = grid_search.fit(xtr, ytr)

```

```python
best_accuracy = gd_sr.best_score_
best_accuracy
```

```python
best_params = gd_sr.best_params_
best_params
```
